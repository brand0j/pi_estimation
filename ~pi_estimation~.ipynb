```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

# Estimate the value of pi by generating random numbers #

def pi_estimate(n): # n is the number of points you wish to generate
    
    points_circle = 0 #used to set up ratio of (points_in_circle/points_total)
    points_total = 0
    
    for _ in range(n):
        
        x=np.random.uniform(-1,1) # generate a random number between -1 & 1
        y=np.random.uniform(-1,1)
        distance = np.sqrt(x**2 + y**2) 
            # since the cirlce is C(0,0) w/radius =1 the distance from a point to (0,0) needs to be less than 1
        
        if distance <=1:
            
            plt.plot(x,y,'o-r',markersize=0.5)# plots points in the circle in red
            points_circle += 1 
            
        else:
            
            plt.plot(x,y,'o-b',markersize=0.5) # plots points outside the circle in blue
            
        points_total += 1
        
    # This is so we can visually see a circle in the plot
    r=1  
    angle = np.linspace(0,2*np.pi,n)
    
    x_c = r*np.cos(angle)
    y_c = r*np.sin(angle)
    plt.plot(x_c,y_c,'k--',linewidth=0.7)
    
    # pi*r^2 = area of circle
    # 2r^2 = area of square
    # (pi*r^2)/2r^2 = (points_circle/points_total) --> (Area of Circle/Area of Square) = 4 * (# of Points in Circle/# of Total Points)

    estimate = 4*(points_circle/points_total)
    plt.title('Estimation of Pi: '+str(estimate))  # our estimation of pi
    plt.grid()
    plt.show()
```

```python
figure(figsize=(10,10))
pi_estimate(10000)
```
